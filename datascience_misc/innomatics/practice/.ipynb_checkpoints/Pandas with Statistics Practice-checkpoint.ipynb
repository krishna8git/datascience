```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
class_df = pd.read_csv('class_groups.csv')
class_df
```

```python
class_df.describe()
```

```python
class_df
```

```python
def plot_groups(data, kde, rug):
    fig, ax = plt.subplots(figsize = (16, 5))
    
    ax1 = plt.subplot(121)
    sns.distplot(data, ax = ax1, rug = rug, kde = kde)

    ax2 = plt.subplot(122)
    ax2.plot(data)
    
    plt.show()
```

```python
plot_groups(class_df.Group1, True, False)
plot_groups(class_df.Group2, True, False)
plot_groups(class_df.Group3, True, False)
plot_groups(class_df.Group4, True, False)
plot_groups(class_df.Group5, True, False)
plot_groups(class_df.Group6, True, False)
```
