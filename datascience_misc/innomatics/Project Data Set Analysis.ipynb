###  Observations

1. The price for few records is zero and in the business this can't be zero. So either we can drop them or treat them.
2. Treating the price: We need to collect the average price of a particular location(where the price is 0.0) and the square feet.

```python
# Returns index positions of those properties whose price is 0.0L

n = d2[d2['price(L)'] == 0].index.tolist()
print(n)
for row in n:
    print("(", d2.loc[row, 'rate_persqft'], ", ", d2.loc[row, 'area_insqft'], " )" )
    
```

```python
# This code multiplies rate per sqft and area and returns the price for the properties
import numpy as np

rows = d2[d2['price(L)'] == 0].index.tolist()
prices = []

for row in rows:
    ratepersqft = d2.loc[row, 'rate_persqft'] # Return sthe value of ratepersqft.
    areainsqft = d2.loc[row, 'area_insqft']
    price = np.round(ratepersqft * areainsqft / 100000, 2)
    d2.loc[row, 'price(L)'] = price
    prices.append(price)
    
```
