```python
import statsmodels.formula.api as sfa
import statsmodels.api as sa
import lmdiag
```

```python
model = sfa.ols(formula = 'Dist~Speed', data = df).fit()
```

```python
print(model.summary())
```

```python
plt.figure(figsize = (15, 15))
lmdiag.plot(model)
plt.show()
```

```python
# Remove outliers and distance and rebuild the model
df_new = df[df.Dist < 120]
df_new.shape
```

```python
model_new = sfa.ols(formula = 'Dist~Speed', data = df_new).fit()
```

```python
print(model_new.summary())
```

```python
plt.figure(figsize = (15, 15))
lmdiag.plot(model_new)
plt.show()
```
