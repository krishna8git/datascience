# 1.3 Pie Chart for Price

```python
f, ax = plt.subplots(figsize = (10, 9)) # Plot Size
ax1 = plt.subplot(211)
f.suptitle('Price percentages, counts and distribution') # plot title
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 1) # Distance b/w each sub pie
interval_1['price(L)'].value_counts(ascending = False).head(10).plot(
    'pie', autopct = "%0.2f%%", explode = explode, ax = ax1)

# Pie plot for top 10 values of price.
ax2 = plt.subplot(223)
interval_1['price(L)'].plot(kind = 'hist') # Histogram for allprices.
ax2.plot()

# ax3 = plt.subplot(224)
# interval_1['price(L)'].plot(kind = 'density' )
```

# Observations:

* Approaximately 19.7% of properties in the price range 0.16 to 1L have price as 10L.
* 13.78% of properties have price as 9.0L
* 4.00% of properties have price as 9.55L
* Number of properties having price ( 8 -10L) are more compared to (0.16 to L)
