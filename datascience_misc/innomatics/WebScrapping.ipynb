```python
import re
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
data = requests.get('https://nano.tatamotors.com/price')
data
```

```python
data_soup = BeautifulSoup(data.text)
data_soup
```

```python
data_tag = data_soup.find_all('figcaption')
data_tag
```

```python
data_srcost = data_soup.find_all('figcaption', class_='showroomCost mb')
data_srcost
```

```python
data_tag = data_soup.find_all('figcaption')
data_tag
```

```python
models = []
engines = []
prices = []

for i in data_tag:
    a = i.text
    
    ############# Model ###########
    model = re.search(r'\b[GE]\w+\s\w+', a)
    if model:# models.append(model)
        models.append(model.group())
    
    ################ Engine ############
    engine = re.search(r'\b[PC]\w+', a)
    if engine:
        engines.append(engine.group())
    
    ################ Price ##############
    price = re.search(r'\b\d+\W\d+\W\d+', a)
    if price:
        prices.append(price.group())
```

```python
print("Models: ", models )
print("Engines: ", engines)
print("Prices: ", prices)
```

```python
nano = pd.DataFrame()
nano['Model'] = models
nano['Engine'] = engines
nano['Price'] = prices
```
