## K-Means Clustering

```python
data = df
data.head()
```

```python
x = data[['Annual Income (k$)', 'Spending Score (1-100)']].values
x
```

```python
from sklearn.cluster import KMeans

wss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i, random_state = 0)
    km.fit(x)
    wss.append(km.inertia_) # with in sum of sqaures values
    # inertia_ is the sum of squared dist from closest cluster

plt.figure(figsize = (18, 8))
plt.plot(range(1, 11), wss, marker = 'o')
plt.title('The Elbow Curve', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wss')
plt.show()
```

```python
km = KMeans(n_clusters=5, random_state = 0)
km.fit(x)
km.inertia_
```

```python
km.cluster_centers_
```

```python
km = KMeans(n_clusters = 5, random_state = 0)
y_means = km.fit_predict(x)
y_means
```

```python
plt.figure(figsize = (18, 8))
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'lime', label = 'miser')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow', label = 'general')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'blue', label = 'target')
plt.scatter(x[y_means == 3, 0], x[y_means == 3, 1], s = 100, c = 'magenta', label = 'spendthrift')
plt.scatter(x[y_means == 4, 0], x[y_means == 4, 1], s = 100, c = 'orange', label = 'careful')

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 200, c = 'red',
           label = 'centroid', marker = '*')

plt.title('K Means Clustering', fontsize = 20)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_means_agg = agg.fit_predict(x)
y_means_agg

plt.figure(figsize = (18, 8))
plt.scatter(x[y_means_agg == 0, 0], x[y_means_agg == 0, 1], s = 100, c = 'lime', label = 'miser')
plt.scatter(x[y_means_agg == 1, 0], x[y_means_agg == 1, 1], s = 100, c = 'yellow', label = 'general')
plt.scatter(x[y_means_agg == 2, 0], x[y_means_agg == 2, 1], s = 100, c = 'blue', label = 'target')
plt.scatter(x[y_means_agg == 3, 0], x[y_means_agg == 3, 1], s = 100, c = 'magenta', label = 'spendthrift')
plt.scatter(x[y_means_agg == 4, 0], x[y_means_agg == 4, 1], s = 100, c = 'orange', label = 'careful')

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 200, c = 'red',
           label = 'centroid', marker = 'x')

plt.style.use('fivethirtyeight')
plt.title('Hierarchial / Agglomerative Clustering', fontsize = 20)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize = (18, 8))
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))

plt.title('Dendrogram', fontsize = 20)
plt.xlabel('Customer')
plt.ylabel('Eculidean Distance')
plt.show()
```
