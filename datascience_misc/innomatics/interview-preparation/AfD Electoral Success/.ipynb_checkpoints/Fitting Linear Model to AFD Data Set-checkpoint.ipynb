##  Feature Engineering
Ideally, We could have looked at the pairplot to get the rough idea of the distribution and correlation of all the features. 
But there are total 151 features in the dataset which makes the visual difficut also takes the long time.

Let's look at the `null` values.

#### Checking for `null` Values.

##### Step1: Check For columns having more than 5 null values and remove them.

```python
# df_afd[df_afd.isnull().sum() > 0]
len(df_afd.columns[df_afd.isnull().sum() >= 5])
```

There are 12 columns that contain more than 5 missing values. I choose to remove them because we have lot columns left in our dataset.

```python
columns_list = df_afd.columns[df_afd.isnull().sum() >= 5].to_list()
columns_list
```

```python
df_afd.drop(columns_list, axis = 1, inplace = True)
```

```python
len(df_afd.columns)
```

##### Step2: Impute Null Values With their mean / median or Mode
*`It's always better to impute with Median as Median is less sensitive to outliers.`*

```python
null_columns_list = df_afd.columns[df_afd.isnull().any()].to_list()
null_columns_list
```

NOTE: First Lets Check the Data Types of these columns

```python
df_afd[null_columns_list].dtypes
```

***`all are numerical, so we can impute it with median.`***

```python
# first lets check the null count for these columns
df_afd[null_columns_list].isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'median')
df_afd[null_columns_list] = imputer.fit_transform(df_afd[null_columns_list])
df_afd[null_columns_list].isnull().sum()
```

    great! No null values. So lets check the whole data set again. 

```python
df_afd.isnull().sum()
```

`Perfect! We now have the cleaner data set. But we only need numerical columns to fit the model for this data. So, let's remove the categorical columns if any.`

```python
df_afd_final = df_afd[df_afd.columns[df_afd.dtypes != 'object']]
```

```python
df_afd_final.shape
```

```python
df_afd_final = df_afd.select_dtypes([np.float64, np.int64])
```

```python
df_afd_final.dtypes.value_counts()
```

```python
df_afd_final.shape
```

Lets look at the pairplot now. We use the `diag_kind = kde` as it shows the distribution of a particular feature along the diagonal.

```python
# First, lets seperate out the X and Y

X = df_afd_final.drop(['vot19_14'], axis = 1)
y = df_afd_final[['vot19_14']]
```

```python
X.columns
```

```python
y.columns
```

```python
X.columns.append(y.columns)
```

```python
for i in range(0, len(X.columns), 5):
    sns.pairplot(data = df_afd_final,
                 x_vars=X.columns[i : i + 5].append(y.columns),
                 y_vars=X.columns[i : i + 5].append(y.columns),
                 diag_kind = 'kde')
```

```python
sns.pairplot(data = df_afd_final,
                 x_vars=X.columns[i : i + 5].append(y.columns),
                 y_vars=X.columns[i : i + 5].append(y.columns),
                 diag_kind = 'kde')
```

## Heatmap for checking multi-collinearity among Independent Dimensions

```python
corr = df_afd_final.corr()
corr
```

from the above heatmap there are lot of independent variables that have collinearity in them. So there is multicoliniarity.

```python
corr.shape[0] // 16
```

```python
grid_size = corr.shape[0] // 16
```

```python
plt.figure(figsize = (20, 15))

plt.subplot(2, 2, 1)
sns.heatmap(corr.iloc[0 : 8, 0 : 8], annot = True, linewidths = 1, cmap="YlGnBu", vmin=-1, vmax=1)

plt.subplot(2, 2, 2)
sns.heatmap(corr.iloc[8 : 16, 8 : 16], annot = True, linewidths = 1, cmap="YlGnBu", vmin=-1, vmax=1)

plt.subplot(2, 2, 3)
sns.heatmap(corr.iloc[16 : 24, 16 : 24], annot = True, linewidths = 1, cmap="YlGnBu", vmin=-1, vmax=1)

plt.subplot(2, 2, 4)
sns.heatmap(corr.iloc[24:32, 24:32], annot = True, linewidths = 1, cmap="YlGnBu", vmin=-1, vmax=1)

plt.show()
```
