Now, if you observe the price columns (Final Price, Slashed Price),  you can see that there is no uniformity in the prices. Every country has their own representations of prices. So we need to convert all of them into Indian Currency INR.

#### 2.6.4 Currency Conversion for `Final Price` and `Slashed Price`

##### 2.6.4.1 Method for converting any Currency to INR

```python
import re
def to_inr(price):
    # First 3 letters is the country code, rest is price value
    print(price)
    country_code = ''
    if price.startswith('₹'):
        country_code = 'INR'
        price = float(price[1:])
    else:
        pricegroup = re.search('([A-z]+)(\d+)',price)
        if pricegroup:
            country_code = pricegroup.group(1)
            price = float(pricegroup.group(2))
        
    conversion_rates = {'NPR' : 0.63,
                        'INR' : 1.0,
                        'AED' : 19.54,
                        'RM' : 17.16,
                        'Rp' : 0.00509,
                        }
    
    if country_code not in conversion_rates.keys():
        return 0.0
    
    return conversion_rates[country_code] * price
```

```python
print(f'{"NPR1250"} = {to_inr("NPR1250")}')
```

```python
print(f'{"Rp115000"} = {to_inr("Rp115000")}')
```

```python
print(f'{"AED223"} = {to_inr("AED223")}')
```

```python
print(f'{"INR223"} = {to_inr("INR223")}')
```

```python
print(f'{"RM276"} = {to_inr("RM276")}')
```

##### 2.6.4.2 Cleaning `Final Price`

```python
oyo_df_cleaned_final_price = oyo_df_cleaned['Final Price'].map(to_inr)
```

#### 2.6.7 Cleaning `Discount(%)`

The discount column is in string format. But we want only the number that represents the percentage. So we need to clean this column:
1. First, check if any null or missing values.
2. We need to remove that `% off` from the value.
3. Then convert the value into int

##### 2.6.7.1. Check if any missing values and Treat them

```python
oyo_df_cleaned['Discount(%)'].isnull().value_counts()
```

No missing values are found from the above observation.

##### 2.6.7.2 Converting `Discount(%)` column into `int`

```python
def discount_to_int(discount_str):
    print(discount_str, "type = ", type(discount_str))
    if discount_str != 0:
        groups = re.search(r'(\d+).+', discount_str)
        return int(groups.group(1))
    else:
        return 0
```

```python
discount_to_int('35% off')
```

```python
oyo_df_cleaned['Discount(%)'] = oyo_df_cleaned['Discount(%)'].fillna(0)
oyo_df_cleaned['Discount(%)'] = oyo_df_cleaned['Discount(%)'].map(discount_to_int)
```
