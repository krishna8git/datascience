```python
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
trans = [ line.rstrip().split(',') for line in open('groceries.csv')]
trans
```

```python
te = TransactionEncoder()
te_transformed = te.fit(trans).transform(trans)
te_transformed
```

```python
int_te_transformed = te_transformed.astype("int")
int_te_transformed
```

```python
df_items = pd.DataFrame(int_te_transformed)
df_items
```

```python
te.columns_
```

```python
df_items.columns = te.columns_
```

```python
te.columns_mapping_
```

```python
df_items
```

```python
import time
x = time.time()
frequent_item_sets = apriori(df_items, min_support = 0.002, use_colnames = True)

y = time.time()

print("Total Time Taken: ", y - x)
```

```python
frequent_item_sets
```

```python
rules = association_rules(frequent_item_sets)
rules
```
