```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# generating the data
X, y = make_circles(n_samples = 1000, noise = 0.05, random_state = 1)
X
```

```python
y
```

```python
colors = ['red', 'green']
plt.scatter(X[:,0], X[:,1], c = y)
plt.show()
```

The kernel trick is to add another dimension to the current data so that even though it isn't linearly separable in the current space, it can be with the addition of another dimension.

```python
X1 = X[:, 0].reshape(-1, 1)
X2 = X[:, 1].reshape(-1, 1)
X3 = (X1 ** 2) + (X2 ** 2)
X = np.hstack((X, X3))
X
```

```python
# visualizing the new data 
fig = plt.figure(figsize = (10, 8))
axes = fig.add_subplot(111, projection = '3d')
axes.scatter (X1, X2, X1 ** 2 + X2 ** 2, c = y, depthshade = True)
plt.show()
```

```python
# fitting SVM classifier
from sklearn import svm
svc = svm.SVC(kernel = 'linear')

svc.fit(X, y)
w = svc.coef_
c = svc.intercept_
print("Coeff:", w)
print("Intercept:", c)
```

```python
# plotting data with the separating hyperplane
x1 = X[:, 0].reshape(-1, 1)
x2 = X[:, 1].reshape(-1, 1)
x1, x2 = np.meshgrid(x1, x2)
x3 = -(w[0][0] * x1 + w[0][1] * x2 + c) / w[0][2]

fig = plt.figure(figsize = (16, 9))
axes2 = fig.add_subplot(111, projection = '3d')
axes2.scatter(X1, X2, X1 ** 2 + X2 ** 2, c = y, depthshade = True)

axes1 = fig.gca(projection = '3d')
axes1.plot_surface(x1, x2, x3, alpha = 0.01)
plt.show()
```
