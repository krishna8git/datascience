```python
def change_quality(quality):
    if quality > 6:
        return 1 # premium
    elif quality > 4: # good
        return 0
    return -1 # average
```

```python
# unit test
print("4 =>", change_quality(4))
print("5 =>", change_quality(5))
print("1 =>", change_quality(1))
print("6 =>", change_quality(6))
print("7 =>", change_quality(7))
```

```python
# apply() or map() anything will do
# df.quality.apply(change_quality)
```

```python
df.quality = df.quality.map(change_quality)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
Cs = [0.001, 0.01, 0.1, 1, 10]
gammas = [0.001, 0.01, 0.1, 1]
kernels = ['rbf', 'linear']
param_grid = {'C': Cs, 'gamma' : gammas, 'kernel' : kernels}
```
