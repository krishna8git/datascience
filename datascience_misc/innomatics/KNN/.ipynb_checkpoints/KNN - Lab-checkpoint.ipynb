```python
norm_df.columns = ['WTT', 'PTI', 'EQW', 'SBI', 'LQE', 'QWG', 'FDJ', 'PJF', 'HQE', 'NXJ']
norm_df.head()
```

```python
norm_df['TARGET CLASS'] = df['TARGET CLASS']
norm_df.head()
```

```python
iv = norm_df.drop('TARGET CLASS', axis = 1)
dv = norm_df['TARGET CLASS']
iv
```

```python
dv
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iv, dv, test_size = 0.3, random_state = 101)
```

# Build the KNN Model with (K = 1)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_1 = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_1.fit(X_train, y_train)
```

    p: integer, optional (default = 2) - Pwer Parameter for the Minkowski Metric.
    When,
    
    p = 1, this is equivalent to using manhatten_distance, 
    euclidean_distance for p = 2

```python
pred = knn_1.predict(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'n_neighbors': np.arange(2,21),
         'p' : [1, 2]}
params
```

```python
knn_cv = KNeighborsClassifier()
model = GridSearchCV(knn_cv, params, cv=7)

#cv = None, to use the default 3/5-fold cross validation,
# integer, to specify the number of folds in a '(Stratified)KFold'
```

```python
import time
start = time.time()
model.fit(X_train, y_train)
end = time.time()
print("Time Taken: ", end - start)
```

```python
model.best_params_
```

```python
error_rate = []
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    predictions_i = knn.predict(X_test)
    error_rate.append(np.mean(predictions_i != y_test))

error_rate
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 21), error_rate, color = 'blue',
        linestyle = '--', marker = 'o', markerfacecolor = 'aqua',
        markersize = 10)
plt.title('Error Rate vs. K-Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```
