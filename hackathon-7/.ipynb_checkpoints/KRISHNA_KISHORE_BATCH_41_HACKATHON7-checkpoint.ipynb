<h2 id="multi_corr">Multi Collinearity</h2>

Since we have too many columns it is not worth trying a heat map with correlation matrix of 281 x 281. Rather than looking at multi-collinearity what we can do is we take every independent variable and find the correlation with the dependent variable and we plot them. The following code does this.

```python
# dependent variable - column no-281
dv = 281
corr = []
for column in df_blog_data.drop([281], axis = 1).columns:
    corr_value = df_blog_data[[column, dv]].corr().iloc[0,1]
    corr.append(corr_value)

half_shape = int(np.sqrt(len(corr) + 1))

IVs = df_blog_data.drop([281], axis= 1)
plt.figure(figsize = (16, 9))
plt.plot(IVs.columns, corr, 'ro-')
plt.title("Correlation Plot for the IVs and the DV")
plt.xlabel("Feature")
plt.ylabel("Correlation Coefficient")
for column in IVs.columns:
    corr_val = corr[column - 1]
    corr_percentage = '{}'.format(round(corr_val, 3))
    if corr_val > 0.5:
        plt.annotate(corr_percentage, xy = (column, corr[column - 1] + 0.02))
    elif corr_val < -0.2:
        plt.annotate(corr_percentage, xy = (column, corr[column - 1] - 0.02))
plt.show()
plt.close()
```

Not much correlation we have b/w dependent variable and independent variables.

<h2 id="feature_engineering">Feature Engineering</h2>

<h3 id="p_values">P-values</h3>

Let's find out the p-values for the correlation b/w every independent variable and the dependent variable just to check whether or not the correlation is `statistically significant`. Lets take the 95% confident, in other words the significance level alpha = 0.05.

```python
from scipy import stats

def p_stats(df, dv):
    """ returns a data frame of pearson correlation and p-value for every feature against the dependent variable.
    
    df The Pandas Data Frame With Independent Features and Dependent Feature
    dv The String which the name of the DV(Dependent Variable)
    """
    # split the data
    x = df.drop([dv], axis = 1)
    y = df[dv]
    
    # find the corrleation b/w every element and the dependent variable
    coeffs = []
    p_values = []
    for column in x.columns:
        # correlation b/w x1, x2, x3, ... xn and y
        pearson_coef, p_value = stats.pearsonr(df[column], y)
        coeffs.append(pearson_coef)
        p_values.append(p_value)
    
    p_stats_df = pd.DataFrame(index = x.columns)
    p_stats_df['Pearson Corr Coeff'] = coeffs
    p_stats_df['P Value'] = p_values
    p_stats_df.index.name = "Columns"
    return p_stats_df
```

```python
target_col = 281 # 281 is the target column
p_stats_df = p_stats(df_blog_data, target_col)
p_stats_df
```

By convention, when the
<ul>
    <li>p-value is $<$ 0.001: we say there is strong evidence that the correlation is significant.</li>
    <li>the p-value is $<$ 0.05: there is moderate evidence that the correlation is significant.</li>
    <li>the p-value is $<$ 0.1: there is weak evidence that the correlation is significant.</li>
    <li>the p-value is $>$ 0.1: there is no evidence that the correlation is significant.</li>
</ul>

From the above output, for all the independent variables the correlation with the dependent variable is statistically significant although the linear relationship is not that strong(as the linear realtionship shown by the pearson's correlation coefficent).


This is good result that all the independent variables seem to be good predictors.


Since we took the 95% confidence, lets take the significance level alpha as 0.05 and check if there are any features that contain less then 0.05.

```python
p_stats_df[p_stats_df['P Value'] < 0.05]
```

There are only 133 columns that have the P-Value which is statistically significant. Lets build our model only with these 133 columns.

```python
selected_features = p_stats_df[p_stats_df['P Value'] < 0.05].index.to_list()
print("Number of selected Features:", len(selected_features))
```

The reason that we took column 10 vs 281, is that when we built our correlation among ivs and dv column no: 10 has the 0.5 correlation but no variable has higher than this. 

```python
# step1: Lets check the distribution of this column
plt.figure(figsize = (15,  5))
plt.subplot(1,2,1)
sns.distplot(x_train[10])

plt.subplot(1,2,2)
sns.regplot(x_train[10], y_train)

plt.show()
plt.close()
```

```python
sx_train = x_train[[10]]
sy_train = y_train

sx_test = df_blog_data_test[[10]]
sy_test = df_blog_data_test[281]

from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(sx_train, y_train)
```

```python
def model_evaluate_dist_plot(actual, predicted):
    """
    This function just plots the distribution of actual and predicted values.
    """
    plt.figure(figsize = (16, 9))

    ax1 = sns.distplot(actual, hist = False, color = "r", label = "Actual Values", )
    sns.distplot(predicted, hist = False, color = "b", label = "Fitted Values", ax = ax1)

    plt.title("Actual Vs Fitted Values")
    plt.xlabel("No of Comments for 24 hours")
    plt.ylabel("Proportion of Comments")
    plt.legend()

    plt.show()
    plt.close()
```

```python
y_hat = lm.predict(sx_test)
model_evaluate_dist_plot(sy_test, y_hat)
```

```python
from sklearn.metrics import mean_squared_error
print("R=Square:", lm.score(sx_test, sy_test))
print("mean_squared_error:", mean_squared_error(sx_test, sy_test))
```

Oops! VERY BAD PREDICTION. Its only a 9% of the variance is explained by the model. Lets leave it and go for ultiple Linear Regression.

<h2 id="model4_dstr">Model #4. Decision Tree Regressor</h2>

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

low, high = 3, 10
hyperparameters= { 'max_depth' : range(low, high)}


regressor = DecisionTreeRegressor(random_state = 101)
clf = GridSearchCV(estimator = regressor, param_grid = hyperparameters, cv = 5)
clf.fit(x_train, y_train)
```

```python
y_hat_dtr = clf.predict(x_test)
model_evaluate_dist_plot(y_test, y_hat_dtr)
```

```python
print("Best Params:", clf.best_params_)
print("Best Estimators:", clf.best_estimator_)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_hat_dtr)
print("Decision Tree Mean Squared Error:", mean_squared_error(y_test, y_hat_dtr))
print("Decision Tree Root Mean Squared Error:", np.sqrt(mse))
```

#### Decision Tree Max Depth Performance

```python
max_depths = range(low, high)

training_scores = []
test_scores = []

for depth in max_depths:
    regressor = DecisionTreeRegressor(max_depth=depth, random_state = 101)
    regressor.fit(x_train, y_train)
    test_scores.append(regressor.score(x_test, y_test))
    training_scores.append(regressor.score(x_train, y_train))
```

```python
# Lets plot the scores against the max_depth
plt.figure(figsize = (10, 7))
ax1 = plt.plot(max_depths, training_scores, 'ro-', label = "Train Data")
plt.plot(max_depths, test_scores, 'bo-', label = "Test Data")
plt.xlabel('Max Depth')
plt.xticks(ticks = range(5, 13))
plt.ylabel('Accuracy')
for i in range(len(max_depths)):
    text_train = '{}'.format(np.round(training_scores[i], 4))
    plt.text(max_depths[i] + 0.1, training_scores[i] + 0.01, text_train)
    
    text_test = '{}'.format(np.round(test_scores[i], 4))
    plt.text(max_depths[i] + 0.1, test_scores[i] - 0.01, text_test)
    
    # print((max_depths[i], training_scores[i], test_scores[i]))

plt.legend()
plt.show()
plt.close()
```

We can see that with the Decision Tree Regressor it is overfitting. Because with train the accuracy is more at the max depth 9 and for test data it is very less.

<h2 id="model5_rfr">Model #5. Random Forest Regressor</h2>

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.externals import joblib


pipeline = make_pipeline(preprocessing.StandardScaler(), RandomForestRegressor(n_estimators=100))
pipeline
```

```python
# Hyper parameters 
hyperparameters= { 'randomforestregressor__max_features' : ['auto', 'sqrt', 'log2'], 
                  'randomforestregressor__max_depth' : [None, 5, 3, 1]}
hyperparameters
```

```python
# Tune Model Using Cross-Validation Pipeline
clf = GridSearchCV(pipeline, hyperparameters, cv=10)
clf.fit(x_train, y_train)
```

```python
# Evaluate Model Pipeline on test data
y_pred_rfr = clf.predict(x_test)
model_evaluate_dist_plot(y_test, y_pred_rfr)
```

```python
print("R2 Score:", r2_score(y_test, y_pred_rfr))
mse = mean_squared_error(y_test, y_pred_rfr)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", np.sqrt(mse))
```

```python
print("Random Forest Regressor Best Parameters:", clf.best_params_)
print("Random Forest Regressor Best Estimators:", clf.best_estimator_)
```

After all the Random Forest Regressor performed best with Max Depth as 5.
