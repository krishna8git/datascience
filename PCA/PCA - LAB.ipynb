```python
import numpy as np
import pandas as pd
```

```python
import re
columns = []
for line in open('Data Desc.txt'):
    # print(line)
    groups = re.match(r'\b(\w+):', line)
    if groups:
        columns.append(groups.group(1))

print(columns)
```

```python
# More Pythonic Way
import re
columns_opt = [re.match(r'\b(\w+):', line).group(1) for line in open('Data Desc.txt') if re.match(r'\b(\w+):', line)]
print(columns_opt)
```

```python
df_train = pd.read_csv('https://query.data.world/s/ksxft7lhmbxpihskwsngwhpuul6lye', header = None)
df_train.columns = columns_opt
df_train
```

```python
df_test = pd.read_csv('https://query.data.world/s/cuqtpuoewpxysusrt5z4igihjah4xo', header = None)
df_test.columns = columns
df_test
```

### Standardizing the Data for PCA

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
SPECT_train_std = scaler.fit_transform(df_train.drop('OVERALL_DIAGNOSIS', axis = 1))
SPECT_test_std = scaler.fit_transform(df_test.drop('OVERALL_DIAGNOSIS', axis = 1 ))
```

```python
print(SPECT_train_std.shape)
print(SPECT_test_std.shape)
```

```python
pd.DataFrame(SPECT_train_std[:5])
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(SPECT_train_std)
```

```python
# The amount of vaiance that each PC explains
var = pca.explained_variance_ratio_
print("Size: ", len(var))
var
```

```python
np.round(var, decimals = 3) * 100
```

```python
# Check if the explained variances sum to 100
np.sum(np.round(var, decimals = 4) * 100)
```

```python
cumsum = np.cumsum(np.round(var, decimals = 4) * 100)
cumsum
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 6))
plt.plot(cumsum, 'ro-')
plt.annotate('22', xy = (22, 95.53))
plt.title('Scree Plot')
plt.xlabel('Number of Variables')
plt.ylabel('Cumulative Sum')
plt.show()
```

```python
# Consider first twenty two components as they are explaining the 95% of variation in the data
pca_model = PCA(n_components = 22)
```

```python
train_pca = pca_model.fit_transform(SPECT_train_std)
```

```python
pca_std_df = pd.DataFrame(train_pca).head()
pca_std_df
```

```python
corr = pd.DataFrame(pca_std_df).corr()

import seaborn as sns
plt.figure(figsize = (16, 9))
sns.heatmap(corr, cmap = 'Blues')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(SPECT_train_std, df_train.OVERALL_DIAGNOSIS)
```

```python
y_actual = df_test.OVERALL_DIAGNOSIS
y_pred = model.predict(SPECT_test_std)
```

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_actual, y_pred))
print("Recall:", metrics.recall_score(y_actual, y_pred))
print("Precision:", metrics.precision_score(y_actual, y_pred))
```

```python
cm = metrics.confusion_matrix(y_actual, y_pred)
print(cm)

sns.heatmap(cm, annot = True, cmap = 'Blues', cbar = False)
```
