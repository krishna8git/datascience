```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe().transpose()
```

```python
df_mini = pd.DataFrame()
x = np.arange(10, 20)
y = 3 * x + 10
df_mini['X'] = x
df_mini['Y'] = y
df_mini
```

```python
sns.regplot(x, y, color='red')
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
model = lm.fit(df_mini[['X']], df_mini.Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[2]])
```

```python
model.predict([[3]])
```

```python
model.predict(df_mini[['X']])
```

```python
df_mini
```

```python
sns.residplot(x, y)
```

```python
sns.regplot(df['highway-mpg'], df['price'])
```

```python
sns.residplot(df['highway-mpg'], df['price'])
```

```python
lm_car_price = LinearRegression()
model_carprice = lm_car_price.fit(df[['highway-mpg']], df.price)
```

```python
model_carprice.coef_
```

```python
model_carprice.intercept_
```

```python
Yhat = model_carprice.predict(df[['highway-mpg']])
```

```python
Yactual = df.price
```

```python
plt.figure(figsize = (10, 6))
sns.distplot(Yactual, hist = False, color = 'RED', label='Actual Values')
sns.distplot(Yhat, hist = False, color = 'BLUE', label = "Predicted Values")
```
